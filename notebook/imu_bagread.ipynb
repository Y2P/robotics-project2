```python
import numpy as np
imu_x = -dffilt2.imu__angular_velocity_x.values.copy()
imu_x -= 0.345464
print(imu_x.shape)
plt.plot(imu_x)
plt.show()
```

```python
imu_cumsum_x = np.cumsum(imu_x * 0.01) 
plt.plot(imu_cumsum_x)
a = np.zeros((len(imu_cumsum_x),1))
rangea = int(len(imu_cumsum_x)/4.0)
a[:rangea] = 0
a[rangea:rangea*2] = math.pi/2.0
a[rangea*2:rangea*3] = math.pi
a[rangea*3:rangea*4] = 1.5 * math.pi
plt.plot(a)
plt.plot(imu_cumsum_x*1.4)
plt.show()
```
